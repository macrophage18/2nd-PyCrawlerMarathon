```python
import requests, json

r = requests.get('https://dcard.tw/_api/posts?popular=true')
data = json.loads(r.text)

n = []
for d in data:
    n.append(d['title'])

print (len(n))
```

```python
for file in data:
    print(file['title'], "\n貼文時間:", file['createdAt'], "留言人數:", file['commentCount'], "按讚人數:", file['likeCount'])
```

```python
#3.計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
Sum1 = 0
Sum2 = 0
com = []
like = []

for comment in data:
    com.append(comment['commentCount'])
    Sum1 = Sum1 + comment['commentCount']

for likeS in data:
    like.append(likeS['likeCount'])
    Sum2 = Sum2 + likeS['likeCount']
    
print('平均留言人數:', Sum1/len(com))
print('平均按讚人數:', Sum2/len(like))
```
